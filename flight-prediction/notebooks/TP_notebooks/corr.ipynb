```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_cleaned = pd.read_feather('../../data/processed/df_cleaned2.feather')
```

```python
correlation_matrix = df_cleaned.corr()
```

```python

plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, mask=mask)
plt.title("Correlation Heatmap")
plt.show()
```

```python


# Calculate the correlation of all features with 'y'
correlation_with_y = df_cleaned.corr()['totalFare'].drop('totalFare')  # Exclude 'Amount' from the correlation list

# List feature names sorted by their correlation values
sorted_feature_names = correlation_with_y.sort_values(ascending=False).index.tolist()
# Print feature names and their correlation scores
print("Features sorted by correlation score with 'y':")
for feature_name in sorted_feature_names:
    correlation_score = correlation_with_y[feature_name]
    print(f"{feature_name}: {correlation_score:.2f}")
```
